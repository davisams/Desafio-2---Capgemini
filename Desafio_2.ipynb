```python
from datetime import datetime
import pandas as pd
import numpy
```

```python

menu = 0
print(''' Escolha a opção e aperte ENTER 
        [1] INFORMAR DADOS
        [2] CONSULTAR DADOS
        [0] SAIR \n''')

      # menu.
menu = int(input())
while menu != 0:
    if menu == 1:

      #Entrada de dados do Cadastro.
      nome_anuncio = str(input('Digite um nome para o anúncio: '))
      cliente = str(input('Digite o nome do Cliente: '))
      data_inicio = input('Digite a data de início. ex: dd/mm/aaaa ')
      data_termino = input('Digite a data de término. ex: dd/mm/aaaa ')
      investimento_dia = input('Digite o valor investido por dia: ')

      # número de dias entre datas.
      data_inicio = datetime.strptime(data_inicio, "%d/%m/%Y")
      data_termino = datetime.strptime(data_termino, "%d/%m/%Y")
      data_inicio_dias = data_inicio.toordinal()
      data_termino_dias = data_termino.toordinal()
      dias = data_termino_dias - data_inicio_dias
      dias = dias + 1

      # total investido.
      total_investido = int (investimento_dia) * int (dias)

      # total de cliques, total de compartilhamentos e total de visualizações.
      valor_investido = round(float(total_investido), 2)

      # visualizações Originais.
      visualizacao_original = int(valor_investido * 30)

      # cliques do anúncio original.
      cliques_anuncio_orginal = int(visualizacao_original * 0.12)

      # primeiro Compartilhamento.
      primeiro_compartilhamento = int(cliques_anuncio_orginal * 0.15)

      # visualizações do primeiro compartilhamento.
      visualizacao_primeiro_compartilhamento = int(primeiro_compartilhamento * 40)

      # cliques do primeiro compartilhamento.
      cliques_primeiro_compartilhamento = int(visualizacao_primeiro_compartilhamento * 0.12)

      # segundo compartilhamento.
      segundo_compartilhamento = int(cliques_primeiro_compartilhamento * 0.15)

      # visualizações do segundo compartilhamento.
      visualizacao_segundo_compartilhamento = int(segundo_compartilhamento * 40)

      # segundo compartilhamento.
      cliques_segundo_compartilhamento = int(visualizacao_segundo_compartilhamento * 0.12)

      # terceiro compartilhamento.
      terceiro_compartilhamento = int(cliques_segundo_compartilhamento * 0.15)

      # visualizações do terceiro compatilhamento.
      visualizacao_terceiro_compartilhamento = int(terceiro_compartilhamento * 40)

      # alcance das visualizações.
      alcance = int(visualizacao_original) + int(visualizacao_primeiro_compartilhamento) + int(visualizacao_segundo_compartilhamento) + int(visualizacao_terceiro_compartilhamento)

      # total de compartilhamentos.
      total_compatilhamentos = int (primeiro_compartilhamento) + int (segundo_compartilhamento) + int (terceiro_compartilhamento)

      # total de cliques.
      total_cliques = int (cliques_anuncio_orginal) + int (cliques_primeiro_compartilhamento) + int (cliques_segundo_compartilhamento)


      # lista do resultado solicitado
      anuncio = {'Total Investido': total_investido, 'Total de Visualizações': alcance, 'Total de cliques': total_cliques, 'Total Investido': total_investido, 'Total de Compartilhamentos': total_compatilhamentos}

      # Salvar dados no arquivo
      banco = open('relatorio.txt', 'a')
      banco.write(str(anuncio) + '\n')
      banco.close()
      print ('Dados registrados')
      break

      # Consulta.
    elif menu == 2:
        pesquisa_cliente = input('Digite o nome do cliente ou data que deseja pesquisar: ')
        arquivo = open('relatorio.txt','r') # Leitura do arquivo para pesquisa dos dados.
        cont =0
        for linha in arquivo:
            linha = linha.rstrip()
            if pesquisa_cliente in linha:
                cont = cont + 1
                print(linha)
        arquivo.close()
        break

      # sair
print('Saindo do sistema.')
print()

      # lista dados
lista = []
lista.append(nome_anuncio)
lista.append(cliente)
lista.append(data_inicio)
lista.append(data_termino)
lista.append(investimento_dia)
print(lista)      

      # lista resultados
anuncio = {'Anúncio':[nome_anuncio],
           'Cliente':[cliente], 
           'Data de Início':[data_inicio],
           'Data de Término':[data_termino], 
           'Investimento Diário':[investimento_dia],
           'Total de Cliques': [total_cliques],
           'Total de Visualizações':[alcance], 
           'Total Investido':[total_investido], 
           'Total de Compartilhamentos':[total_compatilhamentos]}
dataframe = pd.DataFrame(anuncio)
print(dataframe)


```
